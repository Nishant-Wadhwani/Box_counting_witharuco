```python
import cv2.aruco as aruco
import cv2
import matplotlib.pyplot as plt
```

```python
def display(img,cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
aruco_dict_sku = aruco.Dictionary_get(aruco.DICT_7X7_250)  # SKU dictionary
parameters = aruco.DetectorParameters_create()
# Get aruco info
sku_markerLength = 0.05  # 5 cm
pallet_markerLength = 0.05  # 5 cm
aruco_info = [sku_markerLength, pallet_markerLength]
```

```python
frame = cv2.imread('/home/nishant/Wipro/work/box_count_aruco/Output/Mission_1623220553.2488945/MaskedImages/DJI_1059.jpg')
```

```python
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
```

```python
display(gray)
```

```python
cornersS, idsS, rejectedImgPointsS = aruco.detectMarkers(
        gray,
        aruco_dict_sku,
        parameters=parameters)
```

```python
frame = aruco.drawDetectedMarkers(frame, cornersS, idsS)
```
